```python
import os
import re
import collections
import pandas as pd
from numpy import dot
from numpy.linalg import norm
import numpy as np
```

```python
# Define the path to the directory containing the text files
CORPUS_DIR = "C:\RecuperacionInf\ir24a\Taller\data"
documents = {}
```

```python
def clean_text(text):
    # Remover símbolos y paréntesis utilizando expresiones regulares
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text

for filename in os.listdir(CORPUS_DIR):
    if filename.endswith('.txt'):
        file_path = os.path.join(CORPUS_DIR, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read().lower()
            cleaned_text = clean_text(text)
            documents[filename] = cleaned_text
```

# normalización

```python
normalized_word_counts = {}
```

```python
for doc in documents:
    word_count = collections.Counter(documents[doc].split())
    total_words = sum(word_count.values())
    normalized_word_count = {word: count/total_words for word, count in word_count.items()}
    normalized_word_counts[doc] = normalized_word_count
```

```python
df = pd.DataFrame.from_dict(normalized_word_counts)
```

```python
df_no_nan = df.fillna(0)

# Transponer el DataFrame
df_transposed = df_no_nan.transpose()

# Mostrar los primeros 10 registros del DataFrame transpuesto
df_transposed
```

```python
# Creación de Q'
# Obtener los nombres de las columnas
nombres_columnas = df.columns

nombres_columnas
```

```python
# Obtener los valores del DataFrame
valores_df = df_consultas.values

valores_df
```

```python
# Convertir la matriz NumPy a una lista de listas de cadenas
lista_df_valores = valores_df.tolist()

# Crear un DataFrame vacío con las columnas proporcionadas
df = pd.DataFrame(columns=nombres_columnas)

# Iterar sobre las consultas y contar la frecuencia de palabras
for indice_consulta , consulta in enumerate(lista_df_valores):
    conteo_palabras = {}
    for oracion in consulta:
        for palabra in oracion.split():
            if palabra in conteo_palabras:
                conteo_palabras[palabra] += 1
            else:
                conteo_palabras[palabra] = 1
    
    # Agregar los recuentos al DataFrame
    df.loc[indice_consulta ] = [conteo_palabras.get(columna, 0) for columna in nombres_columnas]

df
```
