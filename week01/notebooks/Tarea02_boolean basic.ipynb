```python
import os
import re

# Función para leer un archivo y obtener las palabras
def obtener_palabras(archivo):
    with open(archivo, 'r', encoding='utf-8') as f:
        texto = f.read()
        palabras = re.findall(r'\b\w+\b', texto.lower())  # Expresión regular para encontrar palabras
    return palabras

# Carpeta donde están almacenados los libros
carpeta_libros = '../data/'

# Diccionario para almacenar las palabras de cada libro
diccionario_libros = {}

# Iterar sobre los archivos en la carpeta
for archivo in os.listdir(carpeta_libros):
    if archivo.endswith('.txt'):  
        ruta_completa = os.path.join(carpeta_libros, archivo)
        palabras_libro = obtener_palabras(ruta_completa)
        diccionario_libros[archivo] = palabras_libro

# Mostrar las palabras de cada libro
#for libro, palabras in diccionario_libros.items():
   # print(f'Palabras en {libro}: {palabras}')

# También puedes acceder a las palabras de un libro específico así:
# print(diccionario_libros['nombre_del_libro.txt'])
```

```python
import pandas as pd

# Crear un DataFrame con todas las palabras únicas de todos los libros como índices (filas)
palabras_unicas = set()
for palabras_libro in diccionario_libros.values():
    palabras_unicas.update(palabras_libro)

# Convertir el conjunto de palabras únicas en una lista
palabras_unicas_lista = list(palabras_unicas)

# Crear un DataFrame con las palabras únicas como índices
tabla = pd.DataFrame(index=palabras_unicas_lista)

# Llenar la tabla con True o False según si una palabra está en un libro
for libro, palabras_libro in diccionario_libros.items():
    tabla[libro] = tabla.index.isin(palabras_libro)

tabla_sin_duplicados = tabla.drop_duplicates()

# Mostrar la tabla
#print(tabla)
#print(tabla_sin_duplicados)
tabla_sin_duplicados
```

```python
# Pedir al usuario que ingrese una palabra
palabra_buscar = input("Ingresa la palabra que deseas buscar: ")

# Buscar la palabra en cada columna de la tabla
libros_con_palabra = []
for libro in tabla.columns:
    if palabra_buscar in tabla.index and tabla.loc[palabra_buscar, libro]:
        libros_con_palabra.append(libro)

# Mostrar los libros donde se encuentra la palabra
if libros_con_palabra:
    print(f"La palabra '{palabra_buscar}' se encuentra en los siguientes libros:")
    for libro in libros_con_palabra:
        print(libro)
else:
    print(f"La palabra '{palabra_buscar}' no se encuentra en ningún libro.")


```
